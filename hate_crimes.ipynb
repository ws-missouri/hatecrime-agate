```python
tester = agate.TypeTester(limit=100)
```

```python
#This takes a very long time, FYI.
hate = agate.Table.from_csv('oncampushate111213.csv', column_types=tester)
crime = agate.Table.from_csv('oncampuscrime111213.csv', column_types=tester)
```

```python
print len(hate.rows)
print len(crime.rows)
```

```python
limited = hate.where(
    (lambda row: row['VANDAL11'] > 0) or 
    (lambda row: row['VANDAL12'] > 0) or
    (lambda row: row['VANDAL13'] > 0)
)

c_limited = crime.where(
    (lambda row: row['AGG_A11'] > 0) or
    (lambda row: row['AGG_A12'] > 0) or
    (lambda row: row['AGG_A13'] > 0)
)

hate_allvandal = 0
hate_vandalschool = []
crime_allassault = 0
crime_assaultschool = []

for row in limited.rows: 
    total_vandal = int(row['VANDAL11']) + int(row['VANDAL12']) + int(row['VANDAL13'])
#    print row['INSTNM']
#    print row['BRANCH']
#    print "Student body: " + str(row['Total'])
#    print total_vandal
    hate_vandalschool.append([row['INSTNM'], [row['Total']], total_vandal])
    # print hate_vandalschool
print len(hate_vandalschool)

for row in c_limited.rows: 
    total_assault = int(row['AGG_A11']) + int(row['AGG_A12']) + int(row['AGG_A13'])
    crime_allassault += total_assault
#    print row['INSTNM']
#    print row['BRANCH']
#    print "Student body: " + str(row['Total'])
#    print total_assault
    crime_assaultschool.append([row['INSTNM'], row['Total'], total_assault])
# print crime_allassault
print len(crime_assaultschool)
```

```python
z = float(len(hate.rows))
a = float(len(hate_vandalschool))
b = float(len(crime.rows))
c = float(len(crime_assaultschool))
hv_percent = "{:.0f}%".format(a/z*100)
ca_percent = "{:.0f}%".format(c/b*100)
print "Percent of colleges where hate-related vandalism occured: " + hv_percent
print "Percent of colleges where aggravated assault occured: " + ca_percent
```

```python
print "School name, student body size, number of aggravated assaults 2011-2013"
crime_order = sorted(crime_assaultschool,key=lambda x: x[2],reverse=True)
for i in crime_order:
    print i
```

```python
print "School name, student body size, number of hate-related vandalisms 2011-2013"
hv_order = sorted(hate_vandalschool,key=lambda x: x[2],reverse=True)
for j in hv_order:
    print j
```
